```python
from plotly.offline import init_notebook_mode, iplot, plot
import plotly as py
init_notebook_mode(connected=True)
import plotly.graph_objs as go


category_trace = go.Scatter(
                    x = df.Category,
                    y = df.Sales,
                    mode = "lines",
                    name = "idk",
                    marker = dict(color = 'rgba(56, 117, 9, 0.8)'))
                   

                    
data1 = [category_trace]
layout = dict(title = 'Category vs Sales',
              xaxis= dict(title= 'Categories',ticklen= 5,zeroline= False),
             yaxis =dict(title='Sales'))
fig = dict(data = data1, layout = layout)
iplot(fig)
```

```python
df_Spices.index = pd.to_datetime(df_Spices.index,format="%m-%d-%Y", errors='coerce', dayfirst=True)

sorted_idx = df_Spices.index.sort_values() 
df_Spices = df_Spices.loc[sorted_idx] 
df_Spices
```

```python
Spices_trace = go.Scatter(
                    x = df_Spices.index,
                    y = df_Spices.Sales,
                    mode = "lines",
                    name = "idk",
                    marker = dict(color = 'rgba(56, 117, 9, 0.8)'))
                   

                    
data1 = [Spices_trace]
layout = dict(title = 'Spices sales over time',
              xaxis= dict(title= 'Date',ticklen= 5,zeroline= False),
             yaxis =dict(title='Sales'))
fig = dict(data = data1, layout = layout)
iplot(fig)
```

Every 3 - 5 months
